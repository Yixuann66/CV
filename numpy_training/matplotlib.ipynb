```python
from matplotlib import pyplot as plt

# figsize:指定图的长宽，dpi:图像的清晰度，返回一个图像
# xticks:指定x轴的刻度，rotation:旋转的度数
# yticks:指定y轴的刻度
# xlabel:给x轴添加标签
# ylabel:给y轴添加标签
# plot:绘制折线图
fig = plt.figure(figsize=(5,5),dpi=100)
x = range(2,26,2)

y = [i*i for i in x]
x_label = [str(i) for i in x]
y_label = [str(i) for i in y]
plt.xticks(x,x_label)
plt.yticks(y,y_label)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y)
plt.show()

# 调整x轴刻度
_xtick_labels = [ i for i in list(x)]
plt.xticks(x,_xtick_labels)
plt.yticks(y,y_label)
plt.xlabel('x')
plt.ylabel('y')
# 网格绘制

plt.grid()
plt.plot(x,y)
plt.show()
```

```python
# 绘制多条折线图
x = range(2,26,2)
y1 = [i*i for i in x]
y2 = [i*i*i for i in x]
plt.plot(x,y1,label='y1info')
plt.plot(x,y2,label='y2info')
# plt.figure(figsize=(5,5),dpi=100)
# .legend(prop = {'family':'Microsoft YaHei'})
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 绘制散点
x = range(2,26,2)
y = [i*i for i in x]
plt.scatter(x,y)
plt.show()
```
